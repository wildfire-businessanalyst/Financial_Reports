```python
# Convert 'Acct Code' in Financial to string if not already
Financial['Acct Code'] = Financial['Acct Code'].astype(str)

# Replacing acct name and acct code for consistency.
Financial = Financial.copy()  # Create a copy to avoid modifying the original

# Replace 'Acct Name' values selectively
Financial.loc[Financial['Acct Name'] == 'Sundry Sales - Income', 'Department'] = 'ROSES'
Financial.loc[Financial['Acct Name'] == 'Sundry Sales - Cost', 'Department'] = 'ROSES'
Financial['Acct Name'] = Financial['Acct Name'].replace({
    "Sundry Sales - Income": "Packaging Materials",
    "Sundry Sales - Cost": "Packaging Materials"
})

# Replace 'Acct Code' values selectively (assuming these are numerical columns)
Financial['Acct Code'] = Financial['Acct Code'].replace({"590101":"600301","590105":"600301"})

# Renaming Department names for consistency.
Financial.loc[
    (Financial['Acct Name'] == 'Salaries & Wages') & (Financial['Department'] == 'PACKHOUSE ROSES'),
    'Department'
] = 'ROSES'

Financial.loc[
    (Financial['Acct Name'] == 'Salaries & Wages') & (Financial['Department'] == 'PACKHOUSE HYPERICUM'),
    'Department'
] = 'HYPERICUM'

Financial['Department'] = Financial['Department'].replace({'PACKHOUSE HYPERICUM':'PACKHOUSE', 'PACKHOUSE ROSES':'PACKHOUSE'})

# Create a dictionary mapping AccountCode to Department from Budget
account_to_department = Budget.groupby('AccountCode')['Department'].first().to_dict()

# Fill missing values in Financial's Department column using the mapping
Financial['Department'] = Financial['Department'].fillna(Financial['Acct Code'].map(account_to_department))


# Create a dictionary mapping AccountCode to Category from Budget
account_to_category = Budget.groupby('AccountCode')['Category'].first().to_dict()

# Fill missing values in Financial's Category column using the mapping
Financial['Category'] = Financial['Acct Code'].map(account_to_category)
# Display the updated Financial dataframe
print("Updated Financial Data with filled Department values:")
Financial.head()
```

```python
# Replacing values for datatype conversion
Budget['     BudgetAmount     '] = Budget['     BudgetAmount     '].replace('-', 0)
# Remove commas from 'BudgetAmount' column and convert to numeric
Budget['     BudgetAmount     '] = Budget['     BudgetAmount     '].str.replace(',', '')
Budget.Category.unique()
```

```python
# Attempt conversion to numeric, replacing non-numeric values with zeros
Budget['     BudgetAmount     '] = pd.to_numeric(Budget['     BudgetAmount     '], errors='coerce').fillna(0)
print("Error: Encountered non-numeric values in 'BudgetAmount' column (replaced with zeros).")

# Filtering only the rows that have expenses
All_expenses = Budget.loc[(Budget['Category'] == "DIRECT COST") | (Budget['Category'] == "INDIRECT COST")].copy()

# Group by 'Period' and 'Department', then sum up 'BudgetAmount'
expense_df = All_expenses.groupby(['Period', 'Department'])['     BudgetAmount     '].sum().reset_index()

# Replace 'Category' with 'Expense', and set 'Acct Name' and 'Acct Code'
expense_df['Category'] = 'Expense'
expense_df['AccountName'] = 'Total expense'
expense_df['AccountCode'] = 'TE'

# Reorder the columns to match the required format
expense_df = expense_df[['Department', 'Category', 'AccountCode', 'AccountName', '     BudgetAmount     ', 'Period']]

# Concatenate the original DataFrame with the DataFrame containing the calculated expenses
Budget = pd.concat([Budget, expense_df], ignore_index=True)

# Display the first few rows of the new Budget DataFrame with the 'Expense' category
print("Updated Budget Data with 'Expense' category:")
Budget.loc[Budget['Category'] == "Expense"].tail(10)
```

```python
# Filter only the rows that have income
All_income = Budget.loc[Budget['Category'] == "INCOME"]

# Group by 'Department' and 'Period', then sum up 'BudgetAmount'
income_df = All_income.groupby(['Department', 'Period'])['     BudgetAmount     '].sum().reset_index()
income_df.rename(columns={'     BudgetAmount     ': 'Income'}, inplace=True)

# Filter only the rows that have expenses
All_expenses = Budget.loc[(Budget['Category'] == "DIRECT COST") | (Budget['Category'] == "INDIRECT COST")]

# Group by 'Department' and 'Period', then sum up 'BudgetAmount'
expense_df = All_expenses.groupby(['Department', 'Period'])['     BudgetAmount     '].sum().reset_index()
expense_df.rename(columns={'     BudgetAmount     ': 'Expense'}, inplace=True)

# Merge the income and expense DataFrames on 'Department' and 'Period'
profit_df = pd.merge(income_df, expense_df, on=['Department', 'Period'], how='outer')

# Fill NaN values with 0 for accurate profit calculation
profit_df['Income'].fillna(0, inplace=True)
profit_df['Expense'].fillna(0, inplace=True)

# Calculate profit by subtracting expenses from income
profit_df['Profit'] = profit_df['Income'] - profit_df['Expense']

# Prepare the profit DataFrame for concatenation
profit_df = profit_df[['Department', 'Period', 'Profit']]
profit_df.rename(columns={'Profit': '     BudgetAmount     '}, inplace=True)
profit_df['Category'] = 'PROFIT'
profit_df['AccountCode'] = 'NP'
profit_df['AccountName'] = 'NET PROFIT'

# Ensure the order of columns matches the Budget DataFrame
profit_df = profit_df[['Department', 'Category', 'AccountCode', 'AccountName', '     BudgetAmount     ', 'Period']]

# Drop duplicates if any
profit_df.drop_duplicates(inplace=True)

# Concatenate the original Budget DataFrame with the profit DataFrame
Budget = pd.concat([Budget, profit_df], ignore_index=True)

# Display the updated Budget DataFrame with PROFIT category
print("Updated Budget Data with PROFIT Category:")
print(Budget.loc[Budget['Category'] == "PROFIT"].head())
```

```python
# Filtering only the rows that have income
All_income = Budget.loc[Budget['Category'] == "INCOME"]

# Group by 'Department' and 'Period', then sum up 'BudgetAmount'
income_df = All_income.groupby(['Department', 'Period'])['     BudgetAmount     '].sum().reset_index()
income_df.rename(columns={'     BudgetAmount     ': 'Income'}, inplace=True)

# Filtering only the rows that have Direct Cost
DirectCost = Budget.loc[Budget['Category'] == "DIRECT COST"]

# Group by 'Department' and 'Period', then sum up 'BudgetAmount'
DirectCost_df = DirectCost.groupby(['Department', 'Period'])['     BudgetAmount     '].sum().reset_index()
DirectCost_df.rename(columns={'     BudgetAmount     ': 'Direct Cost'}, inplace=True)

# Merge the income and Direct Cost DataFrames on 'Department' and 'Period'
Gross_profit = pd.merge(income_df, DirectCost_df, on=['Department', 'Period'], how='outer')

# Fill NaN values with 0 for accurate profit calculation
Gross_profit['Income'].fillna(0, inplace=True)
Gross_profit['Direct Cost'].fillna(0, inplace=True)

# Calculate profit by subtracting Direct Cost from Income
Gross_profit['Profit'] = Gross_profit['Income'] - Gross_profit['Direct Cost']

# Prepare the gross profit DataFrame for concatenation
Gross_profit = Gross_profit[['Department', 'Period', 'Profit']]
Gross_profit.rename(columns={'Profit': '     BudgetAmount     '}, inplace=True)
Gross_profit['Category'] = 'GROSS PROFIT'
Gross_profit['AccountCode'] = 'GP'
Gross_profit['AccountName'] = 'GROSS PROFIT'

# Ensure the order of columns matches the Budget DataFrame
Gross_profit = Gross_profit[['Department', 'Category', 'AccountCode', 'AccountName', '     BudgetAmount     ', 'Period']]

# Drop duplicates if any

Gross_profit.drop_duplicates(inplace=True)

# Concatenate the original Budget DataFrame with the gross profit DataFrame
Budget = pd.concat([Budget, Gross_profit], ignore_index=True)

# Display the updated Budget DataFrame with GROSS PROFIT category
print("Updated Budget Data with GROSS PROFIT Category:")
Budget.loc[Budget['Category'] == "GROSS PROFIT"].head()
```

```python
import pandas as pd

# Example Financial DataFrame
Financial = pd.DataFrame({
    'Acct Code': ['101', '102', '103', '104'],
    'Amount Ledger': [1000, 2000, 3000, 4000]
})

# Example Budget DataFrame
Budget = pd.DataFrame({
    'AccountCode': ['101', '102', '102', '104', '105'],
    'Category': ['Expense', 'Income', 'Revenue', 'Expense', 'Expense'],
    'Department': ['A', 'B', 'C', 'A', 'B']
})

# Convert 'Acct Code' in Financial to string
Financial['Acct Code'] = Financial['Acct Code'].astype(str)

# Create a mapping from AccountCode to Category and Department
account_mapping = Budget.groupby('AccountCode').first()[['Category', 'Department']]

# Map 'Acct Code' in Financial to Category and Department using the mapping
Financial['Category'] = Financial['Acct Code'].map(account_mapping['Category'])
Financial['Department'] = Financial['Acct Code'].map(account_mapping['Department'])

# Display the updated Financial DataFrame with Category and Department added
print("Financial Data with Category and Department:")
print(Financial)
```
